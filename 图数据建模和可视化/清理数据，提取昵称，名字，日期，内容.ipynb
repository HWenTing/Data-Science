```python
import json
import pandas as pd
import glob
```

#### 创建清理文件  

```python
# 根据文件名处理数据
def get_key_content(file_name):  

    fp = open(file_name, encoding='utf-8') 
    temp = json.loads(fp.read()) 
    fp.close()
    page = temp['response']['list']  
    poetry = pd.DataFrame(columns=['nick','name','date','content'])
    for post in page:
        nick = post['trackback_author_nick']
        name = post['trackback_author_name']
        date = post['trackback_date'] 
        content = post['content'] 
        poetry = poetry.append({'nick':nick,
                                'name':name,
                                'date':date,
                                'content':content},ignore_index=True)
     
    poetry.to_csv('jkeyContent.csv',index=False,mode='a', header=False,encoding="utf_8_sig")
    poetry.drop(poetry.index,inplace=True)#清空c
def cleanData():
#获取文件夹下csv文件
    csv_files = glob.glob(r'E:/学习/大三/数据科学导论/Twitter/附件/附件2-JPalljson/*.json')
    poetry = pd.DataFrame(columns=['nick','name','date','content'])
    poetry.to_csv('jkeyContent.csv',index=False,encoding="utf_8_sig")   
    print(len(csv_files))
    count=0
    for inputfile in csv_files:
        print(str(count)+'    '+inputfile)
        try:
            get_key_content(inputfile)
        except:
            continue
        count= count+1
```

```python
cleanData()
```

#### 去重

```python
data = pd.read_csv('jkeyContent.csv',lineterminator="\n")
data_unique = data.drop_duplicates()
```

#### 按时间分类

```python
total = len(data_unique)
before = pd.DataFrame(columns=['nick','name','date','content'])
before.to_csv('jbefore.csv',index=False,encoding="utf_8_sig")   
after = pd.DataFrame(columns=['nick','name','date','content'])
after.to_csv('jafter.csv',index=False,encoding="utf_8_sig")  
for i in range(total):
    
    try:
        if(data_unique.iloc[i]['date']<1299822360):#震前
                before = before.append({'nick':data_unique.iloc[i]['nick'],
                                'name':data_unique.iloc[i]['name'],
                                'date':data_unique.iloc[i]['date'],
                                'content':data_unique.iloc[i]['content\r']},ignore_index=True)
        else:
                 after = after.append({'nick':data_unique.iloc[i]['nick'],
                    'name':data_unique.iloc[i]['name'],
                    'date':data_unique.iloc[i]['date'],
                    'content':data_unique.iloc[i]['content\r']},ignore_index=True)  
    except:
        continue
    if(i%1000)==0:
        print(i)
        before.to_csv('jbefore.csv',index=False,mode='a', header=False,encoding="utf_8_sig")
        after.to_csv('jafter.csv',index=False,mode='a', header=False,encoding="utf_8_sig")
        before.drop(before.index,inplace=True)
        after.drop(after.index,inplace=True)

before.to_csv('jbefore.csv',index=False,mode='a', header=False,encoding="utf_8_sig")
after.to_csv('jafter.csv',index=False,mode='a', header=False,encoding="utf_8_sig")
        
```
