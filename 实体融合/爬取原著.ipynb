```python
# coding: utf-8
import re 
import pandas as pd
import requests
from urllib import parse
from urllib import request
import time
header={
    'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/69.0.3497.100 Safari/537.36',
    'Connection':'keep-alive'
}
```

```python
def get_url():    
    url_li = []
    for i in range(9):
        url ='https://www.douban.com/doulist/16224/?start='+str(i*25)+'&sort=time&playable=0&sub_type='
        html = requests.get(url=url,headers=header).content.decode('utf-8')
        movie_item = re.findall(r'<div class="source">(.*?)<img', html,re.S)    
        for i in movie_item:
            url_li=url_li+(re.findall(r'<a href="(.*?)"', i,re.S))
        time.sleep(1)
    return url_li
```

```python
def get_details(url):
    html = requests.get(url=url,headers=header).content.decode('utf-8')

    name = re.findall(r'property="v:itemreviewed">(.*?)<', html,re.S)[0]
    description_raw = re.findall(r'<div class="intro">(.*?)<\/div>', html,re.S)[0]
    li = description_raw.split("<p>")
    description=[]
    for i in li:
        if(len(i.split('</p>'))>1):
            description.append(i.split('</p>')[0])
    description = ''.join(description)
    return name,description
```

```python
def find_book():
    comments = pd.DataFrame(columns=['url','name','description'])
    comments.to_csv('book.csv',index=False,encoding="utf-8")
    url_li = get_url()
    for i in url_li:
        try:
            name,description = get_details(i)
            print(name)
            comments = comments.append({'url':i,'name':name,'description':description},ignore_index=True)
        except:
            print(i)
    comments.to_csv('book.csv',index=False,mode='a', header=False,encoding="utf-8")      
```
